```python
# See all the unique grades

df_inputs_prepr['grade'].unique()
```

```python
# Concatenate this with the good_bad variable 

df1 = pd.concat([df_inputs_prepr['grade'], df_targets_prepr], axis = 1)
df1.head()
```

```python
# We use groupby to find the count of good and bad for the grade

df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].count()
```

```python
# Finding the proportion of good borrowers using the mean 

df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()
```

```python
# We now merge these results into one dataframe

df1 = pd.concat([df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].count(),
                df1.groupby(df1.columns.values[0], as_index = False)[df1.columns.values[1]].mean()], axis = 1)
```

```python
# Removing one of the extra grade columns 

df1 = df1.iloc[: , [0, 1, 3]]
df1
```

```python
# Chenging the names of the columns

df1.columns = [df1.columns.values[0], 'n_obs', 'prop_good']
df1
```

```python
# Calculating the proportion of observations that fall into each grade

df1['prop_n_obs'] = df1['n_obs'] / df1['n_obs'].sum()
```

```python
df1
```

```python
# Calculating the number of good borrowers and bad borrower by grade group 

df1['n_good'] = df1['prop_good'] * df1['n_obs']
df1['n_bad'] = (1 - df1['prop_good']) * df1['n_obs']
df1
```

```python
# Now, we caluclate the porportion of good borrowers and bad borrowes for each grade

df1['prop_n_good'] = df1['n_good']/df1['n_good'].sum()
df1['prop_n_bad'] = df1['n_bad']/df1['n_bad'].sum()
df1
```

```python
# Calculating weight of evidence for variable grade

df1['WoE'] = np.log(df1['prop_n_good'] / df1['prop_n_bad'])
df1
```

```python
# Sorting this result by the weight of evidence

df1 = df1.sort_values(['WoE'])
df1 = df1.reset_index(drop = True)
df1
```

```python
# Differences in weight of evidence 

df1['diff_prop_good'] = df1['prop_good'].diff().abs()
df1['diff_WoE'] = df1['WoE'].diff().abs()
df1
```

```python
# Calculating information value 

df1['IV'] = (df1['prop_n_good'] - df1['prop_n_bad']) * df1['WoE']
df1['IV'] = df1['IV'].sum()
df1

# This variable refers to the information value of the variable grade overall
# The variable grade is found to have a decent predictable power
```

# Automating the IV calculation for other variables

```python
def woe_discrete(df, discrete_variable_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variable_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

# Visualizing the weights of evidence 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
# Plotting variable on the x axis and weight of evidence on the y axis 
# we use a rotation parameter to rotate a label in case they do not fit on the plot 

def plot_by_woe(df_WoE, rotation_of_x_axis_labels = 0):
    x = np.array(df_WoE.iloc[:, 0].apply(str))
    y = df_WoE['WoE']
    plt.figure(figsize = (18,6))
    plt.plot(x,y, marker = 'o', linestyle = '--', color = 'k') # color ='k' makes the color of the data black
    plt.xlabel(df_WoE.columns[0])
    plt.ylabel('Weight of Evidence')
    plt.title('Weight of Evidence by ' + df_WoE.columns[0])
    plt.xticks(rotation = rotation_of_x_axis_labels)
```
